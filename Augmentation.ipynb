```python
import cv2
from keras.preprocessing.image import ImageDataGenerator
import cv2

# Load the image and convert it to grayscale

import numpy as np
from matplotlib import pyplot as plt
exp_frame_augmentation = {
    'rotation_range': 10,
    'width_shift_range': 0.1,
    'height_shift_range': 0.1,
    'shear_range': 0.1,
    'zoom_range': 0.1,
    'horizontal_flip': True,
    'brightness_range': [0.5, 1.5],
}

# Load image
img = cv2.imread('/content/reg_img59.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization

eq_gray = cv2.equalizeHist(gray)

eq_gray = np.expand_dims(eq_gray, axis=-1) # Add a dimension for the single channel

# Apply data augmentation
datagen = ImageDataGenerator(exp_frame_augmentation)
apex_augmented = datagen.apply_transform(eq_gray, exp_frame_augmentation)

# Display original and augmented image
fig, axs = plt.subplots(1, 2, figsize=(10, 10))
axs[0].imshow(eq_gray[:, :, 0], cmap='gray')
axs[0].set_title('Original')
axs[1].imshow(apex_augmented[:, :, 0], cmap='gray')
axs[1].set_title('Augmented')
plt.show()
```

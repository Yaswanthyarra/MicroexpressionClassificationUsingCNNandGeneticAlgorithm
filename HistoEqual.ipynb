```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load image
img = cv2.imread('/content/reg_img59.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display images
fig, axs = plt.subplots(1, 3, figsize=(15,5))
axs[0].imshow(img[:,:,::-1])
axs[0].set_title('Original Image')
axs[1].imshow(gray, cmap='gray')
axs[1].set_title('Grayscale Image')

# Calculate histogram of original grayscale image
hist, bins = np.histogram(gray.ravel(), bins=256, range=[0, 256])

# Display histogram of original grayscale image
axs[2].hist(gray.ravel(), bins=256, range=[0, 256], color='b')
axs[2].set_title('Histogram of Grayscale Image')

# Apply histogram equalization
eq_gray = cv2.equalizeHist(gray)

# Display equalized grayscale image
fig, axs = plt.subplots(1, 2, figsize=(10,5))
axs[0].imshow(eq_gray, cmap='gray')
axs[0].set_title('Equalized Grayscale Image')

# Calculate histogram of equalized grayscale image
eq_hist, eq_bins = np.histogram(eq_gray.ravel(), bins=256, range=[0, 256])

# Display histogram of equalized grayscale image
axs[1].hist(eq_gray.ravel(), bins=256, range=[0, 256], color='r')
axs[1].set_title('Histogram of Equalized Grayscale Image')

plt.show()
```
